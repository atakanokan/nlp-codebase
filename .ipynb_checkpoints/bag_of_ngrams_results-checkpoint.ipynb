```python
import pandas as pd
```

```python
import pickle as pkl
```

```python
results_list = ["param_val_losses_adam_lemma.p",
                "param_val_losses_adam_nolemma.p",
                "param_val_losses_sgd_lemma.p",
                "param_val_losses_sgd_nolemma.p"]
opt_list = ["Adam","Adam","SGD","SGD"]
lemma_list = ["True","False","True","False"]
```

```python
column_list = ["learning rate","n-gram",
               "vocab_size","embedding_size","batch_size"]
column_list.extend(list(range(0,30)))
column_list.extend(["optimizer","lemma"])
print(column_list)
```

```python
new_column_list = column_list[-2:]
new_column_list.extend(column_list[:-2])
print(new_column_list)
```

```python
results = pkl.load(open(results_list[0],"rb"))
print(results)
```

```python
# results_df = pd.DataFrame.from_dict(results).T
# results_df = results_df.reset_index(level=[0,1,2,3,4])
# results_df["optimizer"] = opt_list[0]
# results_df["lemma"] = lemma_list[0]
# results_df.columns = column_list
# results_df = results_df[new_column_list]
# results_df
```

```python
for i in range(0,4):
    #print(i)
    
    results = pkl.load(open(results_list[i],"rb"))
    
    
    if i == 0 :
        results_df = pd.DataFrame.from_dict(results).T
        results_df = results_df.reset_index(level=[0,1,2,3,4])
        results_df["optimizer"] = opt_list[i]
        results_df["lemma"] = lemma_list[i]
        results_df.columns = column_list
        results_df = results_df[new_column_list]
    else:
        new_results_df = pd.DataFrame.from_dict(results).T
        new_results_df = new_results_df.reset_index(level=[0,1,2,3,4])
        new_results_df["optimizer"] = opt_list[i]
        new_results_df["lemma"] = lemma_list[i]
        new_results_df.columns = column_list
        new_results_df = new_results_df[new_column_list]
        
        results_df = results_df.append(new_results_df)

results_df = results_df.reset_index(drop=True)
results_df
```

## Plots

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### All Optimizations

```python
plt.figure(figsize = (10,5))
_ = plt.plot(results_df.iloc[:,7:].T, alpha = 0.2)
```

### Adam vs. SGD

Adam is clearly much faster. More computational resources are needed to see SGD's convergence.

```python
plt.figure(figsize = (10,5))
_ = plt.plot(results_df[results_df["optimizer"] == "Adam"].iloc[:,7:].T,
            color = "red", alpha = 0.2)
_ = plt.plot(results_df[results_df["optimizer"] == "SGD"].iloc[:,7:].T,
            color = "blue", alpha = 0.2)
```

### Different Learning Rates with Adam

```python
plt.figure(figsize = (10,5))
_ = plt.plot(results_df.loc[(results_df["optimizer"] == "Adam") & \
           (results_df["learning rate"] == 1e-2)].iloc[:,7:].T,
            color = "red", alpha = 0.2)
_ = plt.plot(results_df.loc[(results_df["optimizer"] == "Adam") & \
           (results_df["learning rate"] == 1e-1)].iloc[:,7:].T,
            color = "blue", alpha = 0.2)
_ = plt.plot(results_df.loc[(results_df["optimizer"] == "Adam") & \
           (results_df["learning rate"] == 1)].iloc[:,7:].T,
            color = "green", alpha = 0.2)
_ = plt.plot(results_df.loc[(results_df["optimizer"] == "Adam") & \
           (results_df["learning rate"] == 2)].iloc[:,7:].T,
            color = "brown", alpha = 0.2)
```
