```python
import pandas as pd
```

```python
import pickle as pkl
```

```python
results_list = ["param_val_losses_adam_lemma.p",
                "param_val_losses_adam_lemma.p copy",
                "param_val_losses_adam_lemma.p copy 2",
                "param_val_losses_adam_lemma.p copy 3"]
opt_list = ["Adam","Adam","SGD","SGD"]
lemma_list = [True,False,True,False]
```

```python
column_list = ["learning rate","n-gram",
               "vocab_size","embedding_size","batch_size"]
column_list.extend(list(range(0,30)))
column_list.extend(["optimizer","lemmatization"])
print(column_list)
```

```python
new_column_list = column_list[-2:-1]
new_column_list.extend(column_list[:-2])
print(new_column_list)
```

```python
results = pkl.load(open(results_list[0],"rb"))
print(results)
```

```python
for i in range(0,4):
    #print(i)
    
    if i == 0 :
        results_df = pd.DataFrame.from_dict(results).T
        results_df = results_df.reset_index(level=[0,1,2,3,4])
        results_df["optimizer"] = opt_list[i]
        results_df["lemma"] = lemma_list[i]
        results_df.columns = column_list
        results_df = results_df[new_column_list]
    else:
        new_results_df = pd.DataFrame.from_dict(results).T
        new_results_df = new_results_df.reset_index(level=[0,1,2,3,4])
        new_results_df["optimizer"] = opt_list[i]
        new_results_df["lemma"] = lemma_list[i]
        new_results_df.columns = column_list
        new_results_df = new_results_df[new_column_list]
        
        results_df = results_df.append(new_results_df)
        
results_df
```
